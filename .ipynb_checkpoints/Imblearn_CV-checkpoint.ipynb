# Classifying Songs on the Billboard Hot 100

- **True Positive**: Predict song is in the top 10, and it is in the top 10
- **False Positive**: Predict song is in the top 10, but it is not in the top 10
- **True Negative**: Predict song is not in the top 10, and it is not in the top 10
- **False Negative**: Predict song is not in the top 10, but it is in the top 10

- **High Recall**: Most top 10 songs were correctly predicted
    - No concern for number of songs that were incorrectly predicted as top 10
- **High Precision**: Among the predictions of top 10 songs, most predictions are correct
    - No concern for number of missed songs that were in top 10

## Cross-Validation and Upsampling through imblearn



```python
kf = StratifiedKFold(n_splits=5, random_state=42, shuffle=False)
```

```python
# Scale training and test dataframes:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

std = StandardScaler()
std.fit(X_train.values)

X_tr_scaled = std.transform(X_train.values)
# X_val_scaled = std.transform(X_val.values)
X_te_scaled = std.transform(X_test.values)

# Convert arrays back to dataframes - dropped outliers, no dropped p-vals
X_tr = pd.DataFrame(X_tr_scaled, index = X_train.index, columns = X_columns)
# X_v = pd.DataFrame(X_val_scaled, index = X_val.index, columns = X_columns)
X_te = pd.DataFrame(X_te_scaled, index = X_test.index, columns = X_columns)
```

```python
classifiers = [    \
    KNeighborsClassifier(n_neighbors=1),    \
    LogisticRegression(C = 0.12, solver='lbfgs'),    \
    GaussianNB(),    \
    LinearSVC(),    \
    SVC(gamma="auto"),    \
    SVC(kernel="poly", gamma="auto"),    \
    SVC(gamma="scale"),    \
    DecisionTreeClassifier(max_depth=4),    \
    RandomForestClassifier(n_estimators=100),    \
    MLPClassifier(activation="identity"),    \
    MLPClassifier(),    \
    XGBClassifier()]

model_list = ['KNN', 'Logistic Regression', 'Naive Bayes',    \
              'SVM (Linear)', 'SVC', 'SVC (Poly Kernel)', 'SVC (Radial Kernel)',    \
              'Decision Tree', 'Random Forests', 'Neural Network (Linear)',    \
              'Neural Network (Non-linear)',    \
              'XGBoost']

precision_list = []
recall_list = []
f1_score_list = []

for index, classifier in enumerate(classifiers):
    pipe = Pipeline(steps=[('classifier', classifier)])
    imba_pipeline = make_pipeline(SMOTE(random_state=42), classifier)  
    print(classifier)
    precision_list.append(100*cross_val_score(imba_pipeline, X_tr, y_train, scoring='precision', cv=kf))
    recall_list.append(100*cross_val_score(imba_pipeline, X_tr, y_train, scoring='recall', cv=kf))
    f1_score_list.append(100*cross_val_score(imba_pipeline, X_tr, y_train, scoring='f1', cv=kf))
```

```python
precision_avg_list = []
recall_avg_list = []
f1_avg_list = []

for p in precision_list:
    precision_avg_list.append(p.mean())
for r in recall_list:
    recall_avg_list.append(r.mean())
for f in f1_score_list:
    f1_avg_list.append(f.mean())
```

```python
metric_dict = {'Precision': precision_avg_list, 'Recall': recall_avg_list, 'F1 Score': f1_avg_list}
```

```python
metric_df = pd.DataFrame(metric_dict, index = model_list[:-1])
metric_df.sort_values(by='F1 Score', ascending = False, inplace=True)
metric_df.round(1)
```

```python
fig = plt.figure(figsize=[8,8], dpi=100)

# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = metric_df['Precision'].values
bars2 = metric_df['Recall'].values
bars3 = metric_df['F1 Score'].values
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.barh(r1, bars1, color='g', height=barWidth, edgecolor='white', label='Precision')
plt.barh(r2, bars2, color='r', height=barWidth, edgecolor='white', label='Recall')
plt.barh(r3, bars3, color='b', height=barWidth, edgecolor='white', label='F1 Score')
 
# Add xticks on the middle of the group bars
plt.xticks(fontsize=15)
plt.xlabel('Score %', fontsize=15)
plt.yticks([r + barWidth for r in range(len(bars1))], metric_df.index, fontsize=15)
plt.ylabel('Classifier', fontsize=15)
 
# Create legend & Show graphic
plt.legend(fontsize=15)
plt.show()
```

## Cross-validation and Upsampling through Kevin approach

```python
kf_k = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

```python
cv_precision = defaultdict(list)
cv_recall = defaultdict(list)
cv_f1_score = defaultdict(list)
```

```python
X_tr_cv = np.array(X_tr_cv)
y_train_cv = np.array(y_train_cv)
```

```python
model_dict = {'KNN': KNeighborsClassifier(n_neighbors=1),    \
              'Logistic Regression': LogisticRegression(C = 0.12, solver='lbfgs'),    \
              'Naive Bayes': GaussianNB(),    \
              'SVM (Linear)': LinearSVC(),    \
              'SVC': SVC(gamma="auto"),    \
              'SVC (Poly Kernel)': SVC(kernel="poly", gamma="auto"),    \
              'SVC (Radial Kernel)': SVC(gamma="scale"),    \
              'Decision Tree': DecisionTreeClassifier(max_depth=4),    \
              'Random Forests': RandomForestClassifier(n_estimators=100),    \
              'Neural Network (Linear)': MLPClassifier(activation="identity"),    \
              'Neural Network (Non-linear)': MLPClassifier(),    \
              'XGBoost': XGBClassifier()}
```

```python
for name, model in model_dict.items():
    for train_ind, val_ind in kf_k.split(X_tr_cv, y_train_cv):
        X_tr_model, y_tr_model = X_tr_cv[train_ind], y_train_cv[train_ind]
        X_smote_tr, y_smote_train = SMOTE(random_state=42).fit_sample(X_tr_model, y_tr_model)
        X_v_cv, y_val_cv = X_tr_cv[train_ind], y_train_cv[train_ind]
        
        model.fit(X_smote_tr, y_smote_train);
        y_pred_cv = model.predict(X_v_cv)
        
        precision = 100*precision_score(y_val_cv, y_pred_cv)
        recall = 100*recall_score(y_val_cv, y_pred_cv)
        f1 = 100*f1_score(y_val_cv, y_pred_cv)
        
        cv_precision[name].append(precision)
        cv_recall[name].append(recall)
        cv_f1_score[name].append(f1)
        
        print(name, classification_report(y_val_cv, y_pred_cv))
```

```python
precision_avg_dict = {}
recall_avg_dict = {}
f1_avg_dict = {}

for model, p in cv_precision.items():
    precision_avg_dict[model] = sum(p)/len(p)
for model, r in cv_recall.items():
    recall_avg_dict[model] = sum(r)/len(r)
for model, f in cv_f1_score.items():
    f1_avg_dict[model] = sum(f)/len(f)
    
precision_df = pd.DataFrame(precision_avg_dict.values(), index = precision_avg_dict.keys())
precision_df.columns = ['Precision']
recall_df = pd.DataFrame(recall_avg_dict.values(), index = recall_avg_dict.keys())
recall_df.columns = ['Recall']
f1_df = pd.DataFrame(f1_avg_dict.values(), index = f1_avg_dict.keys())
f1_df.columns = ['F1 Score']
metric3_df = precision_df.join(recall_df)
metric3_df = metric3_df.join(f1_df)
```
