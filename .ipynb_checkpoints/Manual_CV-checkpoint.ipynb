# Classifying Songs on the Billboard Hot 100

- **True Positive**: Predict song is in the top 10, and it is in the top 10
- **False Positive (Type I Error)**: Predict song is in the top 10, but it is not in the top 10
- **True Negative**: Predict song is not in the top 10, and it is not in the top 10
- **False Negative (Type II Error)**: Predict song is not in the top 10, but it is in the top 10

- **High Recall**: Most top 10 songs were correctly predicted
    - No concern for number of songs that were incorrectly predicted as top 10
- **High Precision**: Among the predictions of top 10 songs, most predictions are correct
    - No concern for number of missed songs that were in top 10

```python
# # No transformations/changes to data
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# # Remove multicollinear features (only look at seasons)
# X_train, X_test, y_train, y_test = train_test_split(X_logit_seas, y, test_size=0.2, stratify=y, random_state=42)

# Remove multicollinear features (only look at months)
X_train, X_test, y_train, y_test = train_test_split(X_logit_mon, y, test_size=0.2, stratify=y, random_state=42)
```

```python
std = StandardScaler()
std.fit(X_train.values)

X_tr_scaled = std.transform(X_train.values)
X_te_scaled = std.transform(X_test.values)
```

```python
# # No transformations/changes to data
# X_tr = pd.DataFrame(X_tr_scaled, index = X_train.index, columns = X_columns)
# X_tr = sm.add_constant(X_tr)
# X_te = pd.DataFrame(X_te_scaled, index = X_test.index, columns = X_columns)

# # Remove multicollinear features (only look at seasons)
# X_tr = pd.DataFrame(X_tr_scaled, index = X_train.index, columns = X_logit_seas_columns)
# X_tr = sm.add_constant(X_tr)
# X_te = pd.DataFrame(X_te_scaled, index = X_test.index, columns = X_logit_seas_columns)

# Remove multicollinear features (only look at months)
X_tr = pd.DataFrame(X_tr_scaled, index = X_train.index, columns = X_logit_mon_columns)
X_tr = sm.add_constant(X_tr)
X_te = pd.DataFrame(X_te_scaled, index = X_test.index, columns = X_logit_mon_columns)

X_sm_columns = list(X_tr.columns)
```

```python
kf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

```python
cv_precision = defaultdict(list)
cv_recall = defaultdict(list)
cv_f1_score = defaultdict(list)
```

```python
X_tr = np.array(X_tr)
y_train = np.array(y_train)
```

```python
# # No transformations/changes to data
# model_dict = {'KNN': KNeighborsClassifier(n_neighbors=1),    \
#               'Logistic Regression': LogisticRegression(C = 0.12, solver='lbfgs'),    \
#               'Naive Bayes': GaussianNB(),    \
#               'SVM (Linear)': LinearSVC(),    \
#               'SVC': SVC(gamma="auto"),    \
#               'SVC (Poly Kernel)': SVC(kernel="poly", gamma="auto"),    \
#               'SVC (Radial Kernel)': SVC(gamma="scale"),    \
#               'Decision Tree': DecisionTreeClassifier(max_depth=4),    \
#               'Random Forests': RandomForestClassifier(n_estimators=100),    \
#               'Neural Network (Linear)': MLPClassifier(activation="identity"),    \
#               'Neural Network (Non-linear)': MLPClassifier(),    \
#               'XGBoost': XGBClassifier()}

# Remove multicollinear features
model_dict = {'Logistic Regression': LogisticRegression(penalty = 'l2', C = 0.01, solver='lbfgs'),    \
              'Naive Bayes': GaussianNB(),    \
              'LinearSVC': LinearSVC(),    \
              'Decision Tree': DecisionTreeClassifier(max_depth=4),    \
              'Random Forests': RandomForestClassifier(n_estimators=100),    \
              'Neural Network (Linear)': MLPClassifier(activation="identity"),    \
              'XGBoost': XGBClassifier()}
```

```python
for name, model in model_dict.items():
    for train_ind, val_ind in kf.split(X_tr, y_train):
        X_tr_ind, y_tr_ind = X_tr[train_ind], y_train[train_ind]
        X_smote_tr, y_smote_train = SMOTE(random_state=42).fit_sample(X_tr_ind, y_tr_ind)
        
        X_v, y_val = X_tr[val_ind], y_train[val_ind]
        X_smote_v, y_smote_val = SMOTE(random_state=42).fit_sample(X_v, y_val)
        
        model.fit(X_smote_tr, y_smote_train);
        y_pred = model.predict(X_v)
        
        precision = 100*precision_score(y_val, y_pred)
        recall = 100*recall_score(y_val, y_pred)
        f1 = 100*f1_score(y_val, y_pred)
        
        cv_precision[name].append(precision)
        cv_recall[name].append(recall)
        cv_f1_score[name].append(f1)
        
        print(name, classification_report(y_val, y_pred))
```

```python
precision_avg_dict = {}
recall_avg_dict = {}
f1_avg_dict = {}

for model, p in cv_precision.items():
    precision_avg_dict[model] = sum(p)/len(p)
for model, r in cv_recall.items():
    recall_avg_dict[model] = sum(r)/len(r)
for model, f in cv_f1_score.items():
    f1_avg_dict[model] = sum(f)/len(f)
    
precision_df = pd.DataFrame(precision_avg_dict.values(), index = precision_avg_dict.keys())
precision_df.columns = ['Precision']
recall_df = pd.DataFrame(recall_avg_dict.values(), index = recall_avg_dict.keys())
recall_df.columns = ['Recall']
f1_df = pd.DataFrame(f1_avg_dict.values(), index = f1_avg_dict.keys())
f1_df.columns = ['F1 Score']
metric3_df = precision_df.join(recall_df)
metric3_df = metric3_df.join(f1_df)
```

```python
f_beta_list = []

for index, rows in metric3_df.iterrows():
    f_beta = (rows['Precision']*rows['Recall'])/((4*rows['Precision']+rows['Recall'])/(5))
    f_beta_list.append(f_beta)
```

```python
metric3_df['F-beta'] = f_beta_list
```

```python
metric3_df = metric3_df.round(1).sort_values(by='F-beta', ascending=True)
```

- Investment in an artist
    - **\\$500K-\\$2M** (1x-4x)
        - Recording: \\$150K- \\$500K (1x-3.33x)
        - Video: \\$50K- \\$300K (1x-6x)
        - Tour: \\$50K- \\$150K (1x-3x)
        - Marketing: \\$200K- \\$700K (1x-3.5x)
- Music labels move quickly right now
    - They may move on form a song within a week
    - They are not in a position to build a star, but rather refine one
        - ie, investment smaller and shorter term
        - They are signing two new artists per day at a cost of \\$11 million
    - From this perspective, recall is the valued metric
        - Labels are signing many artists (partly to maintain market share), and are putting less money into each
        - If an artist strikes it big, then more attention will be focused on them
        

- Type I Error: Invest in Top 10, don't get expected profit
    - Invest \\$2 million, only get \\$1 million in stream profits (make money back through advance/other resources)
- Type II Error: Don't Invest in Top 10, miss potential profit
    - Invest \\$1 million (expecting \\$1 million profit), miss out on \\$8 million in profit
- Invest in Top 10, get expected profit
    - \\$10 million - \\$2 million = \\$8 million
- Invest in not Top 10, get expected profit
    - \\$2 million - \\$1 million = \\$1 million
___

#### Based on these assumptions, recall is 5 times more important than precision for this business case

```python
cv_precision2 = defaultdict(list)
cv_recall2 = defaultdict(list)
cv_f1_score2 = defaultdict(list)
```

```python
model_dict2 = { #'KNN': KNeighborsClassifier(n_neighbors=1),    \
              'Logistic Regression': LogisticRegression(C = 0.12, solver='lbfgs'),    \
              'Naive Bayes': GaussianNB(),    \
              'SVM (Linear)': LinearSVC(),    \
#               'SVC': SVC(gamma="auto"),    \
#               'SVC (Poly Kernel)': SVC(kernel="poly", gamma="auto"),    \
#               'SVC (Radial Kernel)': SVC(gamma="scale"),    \
              'Decision Tree': DecisionTreeClassifier(max_depth=4),    \
              'Random Forests': RandomForestClassifier(n_estimators=100),    \
              'Neural Network (Linear)': MLPClassifier(activation="identity"),    \
#               'Neural Network (Non-linear)': MLPClassifier(),    \
              'XGBoost': XGBClassifier()}
```

```python
for name, model in model_dict2.items():
    for train_ind, val_ind in kf.split(X_tr, y_train):
        X_tr_ind, y_tr_ind = X_tr[train_ind], y_train[train_ind]
        X_smote_tr, y_smote_train = ADASYN(random_state=42).fit_sample(X_tr_ind, y_tr_ind)
        X_v, y_val = X_tr[val_ind], y_train[val_ind]
        
        model.fit(X_smote_tr, y_smote_train);
        y_pred = model.predict(X_v)
        
        precision = 100*precision_score(y_val, y_pred)
        recall = 100*recall_score(y_val, y_pred)
        f1 = 100*f1_score(y_val, y_pred)
        
        cv_precision2[name].append(precision)
        cv_recall2[name].append(recall)
        cv_f1_score2[name].append(f1)
        
        print(name, classification_report(y_val, y_pred))
```

```python
precision_avg_dict2 = {}
recall_avg_dict2 = {}
f1_avg_dict2 = {}

for model, p in cv_precision2.items():
    precision_avg_dict2[model] = sum(p)/len(p)
for model, r in cv_recall2.items():
    recall_avg_dict2[model] = sum(r)/len(r)
for model, f in cv_f1_score2.items():
    f1_avg_dict2[model] = sum(f)/len(f)
    
precision_df2 = pd.DataFrame(precision_avg_dict2.values(), index = precision_avg_dict2.keys())
precision_df2.columns = ['Precision']
recall_df2 = pd.DataFrame(recall_avg_dict2.values(), index = recall_avg_dict2.keys())
recall_df2.columns = ['Recall']
f1_df2 = pd.DataFrame(f1_avg_dict2.values(), index = f1_avg_dict2.keys())
f1_df2.columns = ['F1 Score']
metric_df = precision_df2.join(recall_df2)
metric_df = metric_df.join(f1_df2)
```
